```python
# 03 â€” Volatility features (Realized Variance) for HAR / HAR-X

**Goal.**  
Build realized volatility features used by HAR and HAR-X:
daily realized variance and its weekly/monthly moving averages.

All feature logic is implemented in `src/commodities_features.py`.
This notebook only documents and executes the feature pipeline.
```

```python
from pathlib import Path
import sys
import pandas as pd

repo = Path("..").resolve()
sys.path.append(str(repo))

from src.commodities_features import build_features_df
```

```python
in_clean = repo / "data" / "processed" / "commodities"
out_feat = repo / "data" / "processed" / "features"
out_feat.mkdir(parents=True, exist_ok=True)

targets = [
    ("gold_clean.csv", "gold_features.csv"),
    ("crude_oil_wti_clean.csv", "crude_oil_wti_features.csv"),
    ("natural_gas_clean.csv", "natural_gas_features.csv"),
]
```

```python
for clean_name, feat_name in targets:
    in_path = in_clean / clean_name
    out_path = out_feat / feat_name

    df = pd.read_csv(in_path)
    df_feat = build_features_df(df, price_col="Price", date_col="Date")
    df_feat.to_csv(out_path, index=False)

    print(clean_name, "=>", feat_name)
```
