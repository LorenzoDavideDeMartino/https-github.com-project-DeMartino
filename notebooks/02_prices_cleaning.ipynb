```python
from pathlib import Path
import sys

# Allow imports from project root
repo = Path("..").resolve()
sys.path.append(str(repo))

from src.data_loader import build_clean_commodity_from_parts
```

```python
raw_commodities = repo / "data" / "raw" / "commodities"
out_clean = repo / "data" / "processed" / "commodities"
out_clean.mkdir(parents=True, exist_ok=True)
```

```python
targets = [
    ("gold", "gold_clean.csv"),
    ("crude_oil_wti", "crude_oil_wti_clean.csv"),
    ("natural_gas", "natural_gas_clean.csv"),
]

for folder, clean_name in targets:
    parts_dir = raw_commodities / folder / "parts"
    out_file = out_clean / clean_name

    build_clean_commodity_from_parts(
        parts_dir=parts_dir,
        out_file=out_file,
    )
```
